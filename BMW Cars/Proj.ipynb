```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

df = pd.read_csv("bmw.csv")
df.head()

```python
df.describe()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.nunique()
```

```python
new = df.corr(numeric_only= True)
sns.heatmap(new , annot= True, fmt = "0.2f")
plt.title("Heatmap")
plt.show()
#We get some relations from here, like Mileage is related to year(-0.77)
#Price is related to year(0.62)
#Price related to engine size(0.46)
#These are some good relation to look into for now
```

```python
sns.lineplot(data = df, x = "year", y = "price", errorbar= None)
plt.xlabel("Price in $")
plt.ylabel("Year")
plt.title("Price v/s Year")
plt.show()
#Prices increase per year
```

```python
sns.countplot(x = "year", data = df)
plt.title("Cars sold per year")
plt.xticks(rotation = 90)
plt.xlabel("Year")
plt.show()
#Yearly Car sales data, shows more cars sold in recent years
```

```python
sns.pairplot(data = df)
plt.show()
#Shows scatterplot of all the variables with one another, we'll take look at some particular ones next
```

```python
temp = df.sort_values("year")
sns.regplot(data=temp, x='year', y='price')
plt.xlabel("Year")
plt.ylabel("Price")
plt.xticks(rotation = 90)
plt.ylim(bottom=0)
plt.show()
#Price increases per year
```

```python
temp = df.sort_values("year")
sns.regplot(data=temp, x='year', y='mileage')
plt.xlabel("Year")
plt.ylabel("Price")
plt.xticks(rotation = 90)
plt.show()
#Mileage decreases per year
```

```python
temp = df.sort_values("year")
sns.regplot(data=temp, x='price', y='mileage')
plt.xlabel("Price")
plt.ylabel("Mileage")
plt.xticks(rotation = 90)
plt.ylim(bottom=0)
plt.show()
#Mileage decreases as Price increases
```

```python
temp = df.sort_values("year")
sns.regplot(data=temp, x='price', y='engineSize')
plt.xlabel("Price")
plt.ylabel("Engine Size")
plt.xticks(rotation = 90)
plt.ylim(bottom=0)
plt.show()
#Engine Size increases as price increases
```

```python
sns.countplot(x = "transmission", data = df)
plt.title("Transmission count")
plt.xlabel("Transmission type")
plt.ylabel("No. of Cars")
plt.show()
```

```python
sns.histplot(df['price'], kde=True)
plt.xlabel("Price")
plt.show()
```

```python
sns.histplot(df['mpg'], kde=True)
plt.xlabel("MPG")
plt.show()
```

```python
sns.histplot(df['engineSize'], kde=True)
plt.xlabel("Engine Size")
plt.show()
```

```python
sns.histplot(df['mileage'], kde=True)
plt.xlabel("Mileage")
plt.show()
```

```python
sns.histplot(df['tax'], kde=True)
plt.xlabel("Tax")
plt.show()
```

## From this Analysis we can say mant variables here are interdependent on each other. We plotted some graphs to try to visualize them, the heatmap and the pairplot alone gives idea on what is going on with this dataset but if we want to look at a particular variable up close we can plot it's graph against our different variables and try to figure out some patterns from there.
