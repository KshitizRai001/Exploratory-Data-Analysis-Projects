```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Diwali Sales Data.csv", encoding='latin1')
df.head()
```

```python
df.describe()
#Has some NaN value have to remove them
```

```python
df.info()
#Seems like the full Column of Status and unnamed 1 is NaN 
```

```python
df.drop(columns = ["Status","unnamed1"], inplace = True)
#Remove them
```

```python
df.head()
```

```python
df.isna().sum()
#Some NaN values in Amount just remove them
```

```python
df.dropna(inplace = True)
```

```python
df.info()
```

```python
df.nunique()
```

```python
new = df.corr(numeric_only= True)
sns.heatmap(new, annot = True, fmt = "0.2f")
plt.title("Heatmap")
plt.show()
#From this heatmap we see that there's no direct corelation
```

```python
sns.histplot(df["Amount"], kde = True)
plt.show()
```

```python
sns.countplot(x = "Gender", data = df)
plt.title("Diwali sales based on gender")
plt.show()
```

```python
sns.countplot(x = "Orders", data = df)
plt.title("Frequency of no. of Orders per person")
plt.show()
```

```python
sns.countplot(x='Age Group', data=df, hue='Gender')
plt.title('Customer count by age group and Gender')
plt.show()
```

```python
sns.countplot(x='Marital_Status', data=df)
plt.title('0 = Married, 1 = Unmarried')
plt.show()
```

```python
sns.barplot(x='Gender', y='Amount', data=df)
plt.title('Total Sales Amount by Gender')
plt.show()
```

```python
sns.barplot(x='Gender', y='Amount', data=df, estimator=sum)
plt.title('Total Sales Amount by Gender')
plt.show()
#Both gender spend the same amount but more females buy stuff so more revenue generated from females
```

```python
plt.figure(figsize=(15, 6))
sns.barplot(x='State', y='Amount', data=df, estimator=sum)
plt.xticks(rotation=90)
plt.title('Total sales by State')
plt.show()
#State-wise Revenue generation
```

```python
plt.figure(figsize=(15, 6))
sns.barplot(x='Occupation', y='Amount', data=df, estimator=sum)
plt.xticks(rotation=90)
plt.title('Total sales by Occupation')
plt.show()
```

```python
plt.figure(figsize=(18, 6))
sns.barplot(x='Product_Category', y='Amount', data=df, estimator=sum)
plt.xticks(rotation=90)
plt.title('Best selling products based on revenue generated')
plt.show()
```

```python
sns.barplot(x = "Zone", y = "Amount", data = df, estimator = sum)
plt.title("Revenue generated based on Zones")
plt.show()
```
