```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("education-economy-data.csv")
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
#NO NA values
```

```python
df.nunique()
```

```python
sns.scatterplot(x = "Unemployment Rate (%)", y = "Literacy Rate",data = df)
plt.show()
```

```python
df.hist(figsize = ( 10, 8))
plt.tight_layout
plt.show()
```

```python
new = df.corr(numeric_only= True)
sns.heatmap(new, annot = True, fmt = '.2f')
plt.title("Heatmap")
plt.show()
#This gives good information
#Physician Density and Literacy Rate (0.61) are related strongly
#Physician Density and GDP per Capita (0.54) are related
#Literacy Rate and GDP per Capita (0.37) are related
#Unemployment Rate and GDP per Capita (-0.20) are negativlely related
#Other values are also related as seen from this heatmap but we'll proceed with these relation for our project
```

```python
sns.regplot(x="Literacy Rate", y="Physician Density", data=df)
plt.title("Physician Density vs. Literacy Rate")
plt.show()
#More Literacy -> More Physician
```

```python
sns.regplot(x="GDP per Capita (Current USD)", y="Physician Density", data=df)

plt.title("Physician Density vs. GDP per Capita")
plt.show()
#More GDP per Capita -> More physician
```

```python
sns.regplot(x="Literacy Rate", y="GDP per Capita (Current USD)", data=df)
plt.title("Literacy Rate vs. GDP per Capita")
plt.show()
```

```python
sns.regplot(x="Unemployment Rate (%)", y="GDP per Capita (Current USD)", data=df)
plt.title("Unemployment Rate vs. GDP per Capita")
plt.show()
#More Literacy rate -> More GDP Per Capita
```
